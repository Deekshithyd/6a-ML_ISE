```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file
df = pd.read_csv(r"C:\Users\sanjay.b.s\Downloads\lahari\cox-violent-parsed_filt.csv")

```

```python
# Set seaborn style
sns.set(style="whitegrid")
```

```python

# Plot 1: Age Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Race Distribution
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='race', order=df['race'].value_counts().index)
plt.title('Race Distribution')
plt.xlabel('Race')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot 3: Risk Score Category vs. Race
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='score_text', hue='race')
plt.title('Risk Score by Race')
plt.xlabel('Risk Score Category')
plt.ylabel('Count')
plt.legend(title='Race', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Plot 4: Prior Count Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['priors_count'], bins=30, kde=True)
plt.title('Distribution of Prior Offenses')
plt.xlabel('Number of Priors')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Plot 5: Age by Risk Score
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='score_text', y='age')
plt.title('Age by Risk Score Category')
plt.xlabel('Risk Score Category')
plt.ylabel('Age')
plt.tight_layout()
plt.show()
```

```python

# Plot 6: Recidivism vs Decile Score
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='is_recid', y='decile_score')
plt.title('Decile Score vs Recidivism')
plt.xlabel('Recidivism (0 = No, 1 = Yes)')
plt.ylabel('Decile Score')
plt.tight_layout()
plt.show()
```
